**Unconstrained optimization**

```
import numpy as np
import matplotlib.pyplot as plt
import random

def f(x,y):
    return (x*x)-(x*y)+(y*y)+(2*x)+(4*y)+3
r1=np.array([0.4657,0.8956,0.3877,0.4902,0.5039])
r2=np.array([0.5319,0.8185,0.8331,0.7677,0.1708])
w=0.90 
c1=1.5
c2=1.5
n=5
t=4
xpos=np.array([-2,-4,2,3,5])
ypos=np.array([-1,2,3,-1,-4])
xvel=np.array([0.23,0.543,0.765,0.326,0.975])
yvel=np.array([0.23,0.567,0.497,0.954,0.134])
xpbest=xpos
ypbest=ypos
gbestx=np.array([5])
gbesty=np.array([-4])

```

**Increasing Number of iteration is increasing the x,y location since it is a unconstrained problem**

```
for i in range(0,1000): 
 for i in range(0,5):
  velx=w*xvel[i]+(c1*r1[i]*(xpbest[i]-xpos[i]))+(c2*r2[i]*(gbestx-xpos[i]))
  vely=w*yvel[i]+(c1*r1[i]*(ypbest[i]-ypos[i]))+(c2*r2[i]*(gbesty-ypos[i]))
  pox=velx+xpos[i]
  poy=vely+ypos[i]
  xvel[i]=velx
  yvel[i]=vely
  xpos[i]=pox
  ypos[i]=poy
  if(f(xpbest[i],ypbest[i])<f(xpos[i],ypos[i])):
   xpbest[i]=xpos[i]
   ypbest[i]=ypos[i]
  if(f(xpbest[i],ypbest[i])>f(gbestx,gbesty)):
   gbestx=xpbest[i]
   gbesty=ypbest[i]
```

**Making it as contrained**

```
for i in range(0,1000): 
 for i in range(0,5):
  velx=w*xvel[i]+(c1*r1[i]*(xpbest[i]-xpos[i]))+(c2*r2[i]*(gbestx-xpos[i]))
  vely=w*yvel[i]+(c1*r1[i]*(ypbest[i]-ypos[i]))+(c2*r2[i]*(gbesty-ypos[i]))
  pox=velx+xpos[i]
  poy=vely+ypos[i]
  if(pox and poy  in range(-4,4)):
   xvel[i]=velx
   yvel[i]=vely
   xpos[i]=pox
   ypos[i]=poy
   if(f(xpbest[i],ypbest[i])<f(xpos[i],ypos[i])):
    xpbest[i]=xpos[i]
    ypbest[i]=ypos[i]
   if(f(xpbest[i],ypbest[i])>f(gbestx,gbesty)):
    gbestx=xpbest[i]
    gbesty=ypbest[i]
```
