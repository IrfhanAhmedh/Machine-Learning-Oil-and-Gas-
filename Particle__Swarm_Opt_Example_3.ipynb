**Example 3**

```
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(8,6))

def f(x,y):
    return -np.cos(x) * np.cos(y) *np. exp(-((x - np.pi)**2 + (y - np.pi)**2))
    
x, y = np.array(np.meshgrid(np.linspace(-5,5,100), np.linspace(-5,5,100)))
z = f(x, y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(x,y,z,500,cmap='plasma')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
plt.show()
```

**Unconstrained optimization**

```
import numpy as np
import matplotlib.pyplot as plt
import random

def f(x,y):
    return -np.cos(x) * np.cos(y) *np. exp(-((x - np.pi)**2 + (y - np.pi)**2))  #Minimize this function
rt=np.array([[0.4657,0.8956,0.3877,0.4902,0.5039],[0.5319,0.8185,0.8331,0.7677,0.1708]],dtype = float)
r=rt.transpose()
w=0.90 
c1=.5
c2=.5
n=5
t=4
xt=np.array([[3.1472,4.0579,-3.7301,4.1338,1.3236],[-4.0246,-2.2150,0.4688,4.5751,4.64589]],dtype= float)
x=xt.transpose()
pbest=x
gbest=np.array([-3.7307,0.4688])
velt=np.array([[0.2194,0.1908,0.3828,0.3976,0.0934],[0.2449,0.2228,0.3232,0.3547,0.3773]],dtype=float)
vel=velt.transpose()
for t in range(0,1000):
 for i in range(0,5):
  vel[i,0]=w*vel[i,0]+[c1*r[i,0]*(pbest[i,0]-x[i,0])]+[c2*r[i,1]*(gbest[0]-x[i,0])]
  vel[i,1]=w*vel[i,1]+[c1*r[i,0]*(pbest[i,1]-x[i,1])]+[c2*r[i,1]*(gbest[1]-x[i,1])]
  x[i,0]=x[i,0]+vel[i,0]
  x[i,1]=x[i,1]+vel[i,1]
 for i in range(0,5):
  if(f(pbest[i,0],pbest[i,1])>f(x[i,0],x[i,1])):
   pbest[i,0]=x[i,0]
   pbest[i,1]=x[i,1]
 for i in range(0,5):
  if(f(pbest[i,0],pbest[i,1])<f(gbest[0],gbest[1])):
    gbest[0]=pbest[i,0]
    gbest[1]=pbest[i,1]
```
