```
x=np.array([2,1,4,1,2])
y=np.array([2,1,4,3,4])
```

**Net Present value**

```
NPV=np.array([[120,130,125,250,350],[145,175,260,300,227],[145,255,215,135,149],[299,175,178,148,160],[145,182,258,188,255]])
plt.imshow(NPV, cmap='Set1', interpolation='nearest')
plt.colorbar()
plt.title("Net Present value at each location")
plt.scatter(x,y,marker='*')
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
import random

def f(x,y):
    return NPV[x,y]  #Minimize this function
rt=np.array([[0.4657,0.8956,0.3877,0.4902,0.5039],[0.5319,0.8185,0.8331,0.7677,0.1708]],dtype = float)
r=rt.transpose()
w=0.90 
c1=.5
c2=.5
n=5
t=4
xt=np.array([[2,1,4,1,2],[2,1,4,3,4]],dtype= int)
x=xt.transpose()
pbest=x
gbest=np.array([1,3])
velt=np.array([[0.2194,0.1908,0.3828,0.3976,0.0934],[0.2449,0.2228,0.3232,0.3547,0.3773]],dtype=float)
vel=velt.transpose()
for t in range (0,5):
 for i in range(0,5):
   vel[i,0]=w*vel[i,0]+[c1*r[i,0]*(pbest[i,0]-x[i,0])]+[c2*r[i,1]*(gbest[0]-x[i,0])]
   vel[i,1]=w*vel[i,1]+[c1*r[i,0]*(pbest[i,1]-x[i,1])]+[c2*r[i,1]*(gbest[1]-x[i,1])]
   x[i,0]=int(x[i,0])+int(vel[i,0]) 
   if(x[i,0]>=4):
    x[i,0]=4
   elif (x[i,0]<=0):
    x[i,0]=0
   x[i,1]=int(x[i,1])+int(vel[i,1])
   if(x[i,1]>=4):
    x[i,1]=4
   elif (x[i,1]<=0):
    x[i,1]=0
 for i in range(0,5):
   if(f(pbest[i,0],pbest[i,1])<f(x[i,0],x[i,1])):
    pbest[i,0]=x[i,0]
    pbest[i,1]=x[i,1]
 for i in range(0,5):
   if(f(pbest[i,0],pbest[i,1])>f(gbest[0],gbest[1])):
    gbest[0]=pbest[i,0]
    gbest[1]=pbest[i,1]
```
